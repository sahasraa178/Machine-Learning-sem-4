```python
import pandas as pd
import numpy as np

#Functions

def compute_mean(matrix):
    return np.mean(matrix, axis=0)

def compute_variance(matrix):
    return np.var(matrix, axis=0)

def compute_std(matrix):
    return np.std(matrix, axis=0)

def compute_centroid(matrix):
    return compute_mean(matrix)

def compute_interclass_distance(centroid1, centroid2):
    return np.linalg.norm(centroid1 - centroid2)

def extract_class_data(x_data, y_data, class_label):
    return x_data[y_data == class_label]

# Main Program

# Loading the  dataset (already locked so that it can specifically extract that particular fetures only be useful of this code)
df_full = pd.read_excel("BERT_Embeddings.xlsx")
x_full = df_full.iloc[:, 0:768].values
y_full = df_full['label'].values

# Choosing two classes from the unique classes from the dataset that is 3
class_a = 1
class_b = 2

# Extracting the data for each class
data_a = extract_class_data(x_full, y_full, class_a)
data_b = extract_class_data(x_full, y_full, class_b)

# Computing centroids
centroid_a = compute_centroid(data_a)
centroid_b = compute_centroid(data_b)

# Computing spreads
std_a = compute_std(data_a)
std_b = compute_std(data_b)

# Computing interclass distance
interclass_dist = compute_interclass_distance(centroid_a, centroid_b)

# Output printing statements
print(f"Class {class_a} centroid (first 5 dims):", centroid_a[:5])
print(f"Class {class_b} centroid (first 5 dims):", centroid_b[:5])
print(f"Class {class_a} standard deviation (first 5 dims):", std_a[:5])
print(f"Class {class_b} standard deviation (first 5 dims):", std_b[:5])
print(f"Interclass distance between class {class_a} and {class_b}:", interclass_dist)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Functions

def minkowski_distance(vec1, vec2, p):
    return sum(abs(a - b) ** p for a, b in zip(vec1, vec2)) ** (1 / p)

def compute_minkowski_distances(vec1, vec2, p_range):
    return [minkowski_distance(vec1, vec2, p) for p in p_range]

def plot_minkowski_distances(p_range, distances):
    plt.plot(p_range, distances, marker='o', color='purple')
    plt.title("Minkowski Distance (p = 1 to 10)")
    plt.xlabel("p value")
    plt.ylabel("Distance")
    plt.grid(True)
    plt.show()

#Main Program

# Load dataset (locked because safe and usefull to specifically extract the features from dataset based on this problem)
df_full = pd.read_excel("BERT_Embeddings.xlsx")
x_full = df_full.iloc[:, 0:768].values
y_full = df_full['label'].values

# Selecting two feature vectors (rows) from dataset
# Feature vector A: Row 0 (index 0), Feature vector B: Row 1 (index 1)
vec_A = x_full[0]
vec_B = x_full[1]

# Computing the Minkowski distances for p = 1 to 10
p_values = list(range(1, 11))
distances = compute_minkowski_distances(vec_A, vec_B, p_values)

# Ploting the distance curve
plot_minkowski_distances(p_values, distances)

# Output results of printing statements
print("Feature vector A: Row 0 from dataset")
print("Feature vector B: Row 1 from dataset")
print("Minkowski distances from p=1 to p=10:")
for p, d in zip(p_values, distances):
    print(f"p = {p}: distance = {d:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance

#Functions

def minkowski_distance_custom(vec1, vec2, p):
    return sum(abs(a - b) ** p for a, b in zip(vec1, vec2)) ** (1 / p)

def minkowski_distance_scipy(vec1, vec2, p):
    return distance.minkowski(vec1, vec2, p)

def compare_minkowski_distances(vec1, vec2, p_range):
    custom_distances = [minkowski_distance_custom(vec1, vec2, p) for p in p_range]
    scipy_distances = [minkowski_distance_scipy(vec1, vec2, p) for p in p_range]
    return custom_distances, scipy_distances

def plot_comparison(p_range, custom_distances, scipy_distances):
    plt.plot(p_range, custom_distances, marker='o', label='Custom', color='blue')
    plt.plot(p_range, scipy_distances, marker='x', label='Scipy', color='green')
    plt.title("Minkowski Distance Comparison")
    plt.xlabel("p value")
    plt.ylabel("Distance")
    plt.legend()
    plt.grid(True)
    plt.show()

# Main Program

# Loading the dataset (locked version because safe and usefull to specifically extract the features from dataset for this problem)
df_full = pd.read_excel("BERT_Embeddings.xlsx")
x_full = df_full.iloc[:, 0:768].values
y_full = df_full['label'].values

# Selecting two feature vectors (rows) from dataset
vec_A = x_full[0]  # Row 0
vec_B = x_full[1]  # Row 1

# Comparing distances for p = 1 to 10
p_values = list(range(1, 11))
custom_dists, scipy_dists = compare_minkowski_distances(vec_A, vec_B, p_values)

# Ploting the comparison
plot_comparison(p_values, custom_dists, scipy_dists)

# Output results printing statements
print("Feature vector A: Row 0 from dataset")
print("Feature vector B: Row 1 from dataset")
print("Comparison of Minkowski distances (Custom vs Scipy):")
for p, d1, d2 in zip(p_values, custom_dists, scipy_dists):
    print(f"p = {p}: Custom = {d1:.4f}, Scipy = {d2:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from collections import Counter

#Custom kNN Function

def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def knn_predict_custom(x_train, y_train, x_test, k=3):
    predictions = []
    for test_point in x_test:
        distances = [euclidean_distance(test_point, train_point) for train_point in x_train]
        k_indices = np.argsort(distances)[:k]
        k_labels = [y_train[i] for i in k_indices]
        most_common = Counter(k_labels).most_common(1)[0][0]
        predictions.append(most_common)
    return np.array(predictions)

#Load and Prepare Data

df = pd.read_excel("BERT_Embeddings.xlsx")
x = df.iloc[:, 0:768].values
y = df['label'].values

# Filtering two classes
mask = (y == 1) | (y == 2)
x_two = x[mask]
y_two = y[mask]

# Spliting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x_two, y_two, test_size=0.3, random_state=42)

#Package kNN
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
y_pred_package = model.predict(X_test)

#Custom kNN
y_pred_custom = knn_predict_custom(X_train, y_train, X_test, k=3)

#Comparing the Results
matches = np.sum(y_pred_package == y_pred_custom)
total = len(y_test)
print("Matching predictions:", matches, "out of", total)
print("Accuracy of package kNN:", round(model.score(X_test, y_test) * 100, 2), "%")

#finding the Accuracy of custom kNN
custom_accuracy = np.sum(y_pred_custom == y_test) / total
print("Accuracy of custom kNN:", round(custom_accuracy * 100, 2), "%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Loading and Prepare Data
df = pd.read_excel("BERT_Embeddings.xlsx")
x = df.iloc[:, 0:768].values
y = df['label'].values

# Filtering only two classes (1 and 2)
mask = (y == 1) | (y == 2)
x_two = x[mask]
y_two = y[mask]

# Spliting into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_two, y_two, test_size=0.3, random_state=42)

# Compare k = 1 and k = 3
model_k1 = KNeighborsClassifier(n_neighbors=1)
model_k1.fit(X_train, y_train)
acc_k1 = model_k1.score(X_test, y_test)

model_k3 = KNeighborsClassifier(n_neighbors=3)
model_k3.fit(X_train, y_train)
acc_k3 = model_k3.score(X_test, y_test)

print("Accuracy with k = 1 (NN):", round(acc_k1 * 100, 2), "%")
print("Accuracy with k = 3 (kNN):", round(acc_k3 * 100, 2), "%")

#Accuracy Plot for k = 1 to 11
k_values = list(range(1, 12))
accuracies = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    accuracies.append(acc)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='blue')
plt.title("Accuracy vs k (Number of Neighbors)")
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Function for computing confusion matrix for binary classification
def compute_confusion_matrix(y_true, y_pred):
    labels = np.unique(y_true)
    matrix = np.zeros((2, 2), dtype=int)  # [[TN, FP], [FN, TP]]
    for true, pred in zip(y_true, y_pred):
        i = int(true == labels[1])  # 0 if class1, 1 if class2
        j = int(pred == labels[1])
        matrix[i][j] += 1
    return matrix

# Function for computing accuracy, precision, recall, and F-score
def compute_metrics(conf_matrix, beta=1):
    TN, FP = conf_matrix[0]
    FN, TP = conf_matrix[1]

    accuracy = (TP + TN) / np.sum(conf_matrix)
    precision = TP / (TP + FP) if (TP + FP) else 0
    recall = TP / (TP + FN) if (TP + FN) else 0
    if precision + recall == 0:
        f_beta = 0
    else:
        f_beta = (1 + beta**2) * (precision * recall) / ((beta**2 * precision) + recall)

    return accuracy, precision, recall, f_beta

# The usage on test data
y_pred_test = model.predict(X_test)
cm_test = compute_confusion_matrix(y_test, y_pred_test)
acc, prec, rec, f1 = compute_metrics(cm_test, beta=1)

# Output results printing statements
print("Confusion Matrix (Test):\n", cm_test)
print(f"Accuracy: {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print(f"F1-Score: {f1:.4f}")
```
